```python
#imports

import pandas as pd
import numpy as np


#CLF tree imports
from sklearn import tree
import pydotplus 
from IPython.display import Image
from sklearn.metrics import accuracy_score

#knn imports
from sklearn.neighbors import KNeighborsClassifier
from sklearn.utils import shuffle
```

```python
#pre proccessing for training data
dateColumns=['Number of frequent flyer miles earned per year', 'Percentage of time spent playing video games', 'Liters of ice cream consumed per week','Class'] 
dateTrainData = pd.read_csv('datingData_training.txt', delimiter=r"\s+", names=dateColumns)


#preprocess the data, change classes to 0 1 2

dateTrainData.replace('largeDoses', 0, inplace=True)
dateTrainData.replace('smallDoses', 1, inplace=True)
dateTrainData.replace('didntLike', 2, inplace=True)


dateTrainData
```

```python
#pre proccessing for training data
dateColumns=['Number of frequent flyer miles earned per year', 'Percentage of time spent playing video games', 'Liters of ice cream consumed per week','Class'] 
dateTestData = pd.read_csv('datingData_test.txt', delimiter=r"\s+", names=dateColumns)


#preprocess the data, change classes to 0 1 2

dateTestData.replace('largeDoses', 0, inplace=True)
dateTestData.replace('smallDoses', 1, inplace=True)
dateTestData.replace('didntLike', 2, inplace=True)


dateTestData
```

```python
#date data tree


repeats = 1
accuracies = []
for i in range(repeats):
    #set data
    Y = dateTrainData['Class']
    X = dateTrainData.drop(['Class'],axis=1)
    testY = dateTestData['Class']
    testX = dateTestData.drop(['Class'],axis=1)

    #make tree
    clf = tree.DecisionTreeClassifier(criterion='entropy',max_depth=3)
    clf = clf.fit(X, Y)
    
    # get accuracy
    predY = clf.predict(testX)
    testY = dateTestData['Class']
    acc = accuracy_score(testY, predY)
    accuracies.append(acc) 
    print('Accuracy on test number ', i+1, ' is %.2f' % (acc))

#avg = accuracies[0] +accuracies[1] +accuracies[2] +accuracies[3] +accuracies[4]
avg = 0
for a in accuracies:
    avg = a + avg
print('average accuracy is: ', avg/repeats)


#this is the code that prints out a tree thing
#dot_data = tree.export_graphviz(clf, feature_names=X.columns, class_names=['0','1','2'], filled=True, 
                                #out_file=None) 
#graph = pydotplus.graph_from_dot_data(dot_data) 
#Image(graph.create_png())
```

```python
#KNN algorithm for dating data

#number of runs
repeats = 1
accuracies = []
for i in range(repeats):
    
    Y = dateTrainData['Class']
    X = dateTrainData.drop(['Class'],axis=1)
    testY = dateTestData['Class']
    testX = dateTestData.drop(['Class'],axis=1)

   #make the KNN classifier
    knn = KNeighborsClassifier(n_neighbors=10, metric='minkowski', p=2)
    knn.fit(X, Y)

    # get accuracy
    predY = knn.predict(testX)
    acc = accuracy_score(testY, predY)
    accuracies.append(acc) 
    print('Accuracy on test number ', i+1, ' is %.2f' % (acc))


avg = 0
for a in accuracies:
    avg = a + avg
print('average accuracy is: ', avg/repeats)
```
